CLASSIFICATION THROUGH S.V.M 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
```

```python
file = pd.read_csv("E:/Tree_Classification/classification-of-Trees.csv")
file
```

```python
df_names = file
```

```python
xfeatures = df_names["names"]
```

```python
#Feature_Extraction
cv = CountVectorizer()
x =  cv.fit_transform(xfeatures)
```

```python
cv.get_feature_names()
```

```python
y = df_names.fruitful_nonfruitful
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.5,random_state=75)
```

```python
from sklearn.svm import SVC
clf = SVC(C=1.0, kernel="linear", degree=3, gamma=0.1)
```

```python
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
clf.score(x_train,y_train)
```

```python
list1 = []
with open("E:/Tree_Classification/sent.txt", encoding = 'utf-8') as f:
    data = f.readlines()
data
```

```python
for values in data:
    list1.append(values)
list1
```

```python
list_2 = [entry.strip('\n') for entry in list1]
list_2
```

```python
#Sample Prediction
vec = cv.transform(list_2).toarray()
```

```python
#Prediction01
pr = clf.predict(vec)
pr
```

```python
#Prediction02
def pred(a):
    vect = cv.transform(a).toarray()
    if clf.predict(vect) == 0:
        print("Fruitless")
    elif clf.predict(vect) == 1:
        print("Fruitful")
    else:
        print("Invalid name Entry")
        
x = ["مہندی کا درخت بھارت میں کثرت سے پایا جاتا ہے"]
pred(x)
```

```python
from sklearn.metrics import confusion_matrix
results = confusion_matrix(y_test,y_train) 
print('Confusion Matrix :')
print(results)
```
